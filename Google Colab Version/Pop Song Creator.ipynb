#Preprocess Data
**Compile all lyrics into one dataset and check the unique characters. Write method to encode and decode characters.**

```python
#Combine all lyrics 
compiled_array = []

for song in glob.glob(os.path.join(songs_folder_loc, '*.txt')):
  with open(song, 'r', encoding='utf-8-sig') as file:
    compiled_array.append(file.read())

compiled_songs = '/n'.join(compiled_array)
```

```python
#Check total length of compiled songs
len(compiled_songs)
```

```python
#See all unique characters
charset = sorted(set(compiled_songs))
charset
```

```python
#Count total number of non-English characters
position = charset.index('}') + 1
non_eng_count = 0
for char in charset[position:]:
  non_eng_count += compiled_songs.count(char)

print('Total non-English characters : '+ str(non_eng_count))
print('Percentage of non-English characters : '+ str(round(100*non_eng_count/len(compiled_songs),4)) + '%')
```

```python
#Create method to encode and decode characters
char_to_index = {char:index for index, char in enumerate(charset)}
index_to_char = np.array(charset)
```

```python
#Test methods
print(char_to_index['Z'])
print(index_to_char[59])

print(char_to_index['z'])
print(index_to_char[89])
```

```python
#Encode entire song file to numbers
encoded_songs = np.array([char_to_index[char] for char in compiled_songs])
```

#Batch Creation
**Prepare dataset for model fitting. Set training sequence length followed by batch size.**

```python
#Convert into Dataset for easy manipulation later during model fitting
encoded_dataset = tf.data.Dataset.from_tensor_slices(encoded_songs)
```

```python
#Set a sequence of characters for model to train on
#Take a line of lyrics as the sequence and check in total how many sequences would there be
seq_len = 60
total_num_seq = len(encoded_songs) // seq_len
total_num_seq
```

```python
#Create training sequence
sequences = encoded_dataset.batch(seq_len, drop_remainder=True)
```

```python
#Create function for "moving" consecutive sequences for 'map' function later on
def moving_seq(seq):
  current_seq = seq[:-1]
  next_seq = seq[1:]

  return current_seq, next_seq
```

```python
#Convert the sequences to "moving" using 'map'
moving_seq = sequences.map(moving_seq)
```

```python
#Check length of sequences
len(moving_seq) == total_num_seq
```

```python
#Check sequences are now moving character by character
for current, next in moving_seq.take(1):
  print(current.numpy())
  print(''.join(index_to_char[current]))
  print(next.numpy())
  print(''.join(index_to_char[next]))
```

```python
from tensorflow.python.ops.gen_dataset_ops import shuffle_dataset
#Create training batches of moving sequences.
batch_size = 128
data_batches = moving_seq.shuffle(buffer_size=10000).batch(batch_size, drop_remainder=True)
```

#Model Creation and Fitting
**Apply DeepMoji architecture with one Embedding layer, one GRU layer and one Dense layer. Fit model to dataset.**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense,Embedding,Dropout,GRU
from tensorflow.keras.losses import sparse_categorical_crossentropy
from tensorflow.keras.models import load_model
```

```python
#Setting the parameters
char_size = len(charset)
print(len(charset))
embed_dim = 64
rnn_neurons = 1024 #arbituary
```

```python
def sparse_cat_loss(y_true, y_pred):
  return sparse_categorical_crossentropy(y_true, y_pred, from_logits=True) 
```

```python
def create_model(char_size, embed_dim, rnn_neurons, batch_size):
  model = Sequential()
  model.add(Embedding(char_size,embed_dim,batch_input_shape=[batch_size,None]))
  model.add(GRU(rnn_neurons, return_sequences=True, stateful=True,recurrent_initializer='glorot_uniform'))
  model.add(Dense(char_size))
  model.compile(optimizer='adam', loss=sparse_cat_loss)

  return model
```

```python
model = create_model(char_size, embed_dim, rnn_neurons, batch_size)
model.summary()
```

```python
epochs = 30
model.fit(data_batches, epochs=epochs)
```

```python
model.save('./data/MyDrive/AI Python Portfolio/(NLP) Pop Song Creator/pop_song_creator_model.h5')
```

#Song Creation
**Create new model with batch size of one for prediction. If not, a batch size of 128 (or whatever batch_size value was trained on previously) needs to be suppled to the model for prediction. Load trained model's weights to the new model and create a "new song"by predicting the characters.**



```python
model_with_one_batch = create_model(char_size, embed_dim, rnn_neurons, batch_size=1)
model_with_one_batch.set_weights(model.get_weights())
model_with_one_batch.build(tf.TensorShape([1,None]))
model_with_one_batch.summary()
```

```python
#Create prediction function which uses model to predict and return the characters
def create_new_song(model, start_char, char_to_generate=2000,temp=1.0):
  song_created = []

  encoded_start_char = [char_to_index[char] for char in start_char]
  #Reshape input, adding one more dimensions. 
  encoded_start_char = tf.expand_dims(encoded_start_char, 0) #Previously it is []. After 'tf.expand_dims' becomes [[]]

  #Previously while training 'stateful'was set to True. This is to clear the stats stored.
  model.reset_states()

  for char in range(char_to_generate):
    pred = model(encoded_start_char)
    pred = tf.squeeze(pred, 0) #remove one dimension
    pred = pred / temp #adjusted to temperature
    #[1,0] Selects from the batch (-1 indicates same as what is passed in. 
    #In this case there is only 1 batch so it also works if the value is set to 0) and 
    #the index of the prediction (0 means the first of the X num_samples chosen. 
    #In this case we only select 1 num_sample so 0 is the only option.) extracted.
    converted_probabilities_to_charindex = tf.random.categorical(pred, num_samples=1)[-1, 0].numpy() 
    encoded_start_char = tf.expand_dims([converted_probabilities_to_charindex],0) #expand [55] to [[55]] to match [batch, index] shape of the argument to model() i.e. encoded_start_char.
    song_created.append(index_to_char[converted_probabilities_to_charindex]) #takes in [55] (not expanded) and append to prediction list giving ['r', 't', 'p']. ie. song_created. 


  return (start_char + ''.join(song_created))
```

```python
print(create_new_song(model_with_one_batch,"Happy"))
```
